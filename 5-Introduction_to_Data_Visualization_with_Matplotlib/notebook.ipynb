# Introduction to data visualization with Matplotlib

### Datasets

```python
import pandas as pd
import matplotlib.pyplot as plt
seattle_weather_stations = pd.read_csv('datasets/seattle_weather.csv', index_col=2)
austin_weather_stations = pd.read_csv('datasets/austin_weather.csv', index_col=2)
seattle_weather = seattle_weather_stations.query('STATION == "USW00094290"')
seattle_weather['MONTH'] = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
austin_weather = austin_weather_stations.query('STATION == "USW00013904"')
austin_weather['MONTH'] = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
climate_change = pd.read_csv('datasets/climate_change.csv', parse_dates=['date'], index_col='date')
medals = pd.read_csv('datasets/medals_by_country_2016.csv')
medalist_weights = pd.read_csv('datasets/summer2016.csv')
mens_rowing = medalist_weights.query('Sport == "Rowing" and Sex == "M"')
mens_gymnastics = medalist_weights.query('Sport == "Gymnastics" and Sex == "M"')
```

### Creating small multiples with plt.subplots
Instructions:
<ul>
<li>Create a Figure and an array of subplots with 2 rows and 2 columns.</li>
<li>Addressing the top left Axes as index 0, 0, plot the Seattle precipitation.</li>
<li>In the top right (index 0,1), plot Seattle temperatures.</li>
<li>In the bottom left (1, 0) and bottom right (1, 1) plot Austin precipitations and temperatures.</li>

```python
# Create a Figure and an array of subplots with 2 rows and 2 columns
fig, ax = plt.subplots(2, 2)

# Addressing the top left Axes as index 0, 0, plot month and Seattle precipitation
ax[0, 0].plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'])

# In the top right (index 0,1), plot month and Seattle temperatures
ax[0, 1].plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'])

# In the bottom left (1, 0) plot month and Austin precipitations
ax[1,0].plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-NORMAL'])

# In the bottom right (1, 1) plot month and Austin temperatures
ax[1,1].plot(austin_weather['MONTH'], austin_weather['MLY-TAVG-NORMAL'])
plt.show()
```

### Small multiples with shared y axis
Instructions:
<ul>
<li>Create a Figure with an array of two Axes objects that share their y-axis range.</li>
<li>Plot Seattle's "MLY-PRCP-NORMAL" in a solid blue line in the top Axes.</li>

### Defining a function that plots time-series data
Instructions:
<ul>
<li>Define a function called plot_timeseries that takes as input an Axes object (axes), data (x,y), a string with the name of a color and strings for x- and y-axis labels.</li>
<li>Plot y as a function of in the color provided as the input color.</li>
<li>Set the x- and y-axis labels using the provided input xlabel and ylabel, setting the y-axis label color using color.</li>
<li>Set the y-axis tick parameters using the tick_params method of the Axes object, setting the colors key-word to color.</li>
</ul>

```python
# Define a function called plot_timeseries
def plot_timeseries(axes, x, y, color, xlabel, ylabel):

  # Plot the inputs x,y in the provided color
  axes.plot(x, y, color=color)

  # Set the x-axis label
  axes.set_xlabel(xlabel)

  # Set the y-axis label
  axes.set_ylabel(ylabel, color=color)

  # Set the colors tick params for y-axis
  axes.tick_params('y', colors=color)
```

### Using a plotting function
Instructions:
<ul>
<li>In the provided ax object, use the function plot_timeseries to plot the "co2" column in blue, with the x-axis label "Time (years)" and y-axis label "CO2 levels".</li>
<li>Use the ax.twinx method to add an Axes object to the figure that shares the x-axis with ax.</li>
<li>Use the function plot_timeseries to add the data in the "relative_temp" column in red to the twin Axes object, with the x-axis label "Time (years)" and y-axis label "Relative temperature (Celsius)".</li>
</ul>

```python
fig, ax = plt.subplots()

# Plot the CO2 levels time-series in blue
plot_timeseries(ax, climate_change.index, climate_change['co2'], "blue", "Time (years)", "CO2 levels")

# Create a twin Axes object that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], "red", "Time (years)", "Relative temperature (Celsius)")

plt.show()
```

### Annotating a plot of time-series data
Instructions:
<ul>
<li>Use the ax.plot method to plot the DataFrame index against the relative_temp column.</li>
<li>Use the annotate method to add the text '>1 degree' in the location (pd.Timestamp('2015-10-06'), 1).</li>
</ul>

```python
fig, ax = plt.subplots()

# Plot the relative temperature data
ax.plot(climate_change.index, climate_change['relative_temp'])

# Annotate the date at which temperatures exceeded 1 degree
ax.annotate('>1 degree', (pd.Timestamp('2015-10-06'), 1))

plt.show()
```

### Plotting time-series: putting it all together
Instructions:
<ul>
<li>Use the plot_timeseries function to plot CO2 levels against time. Set xlabel to "Time (years)" ylabel to "CO2 levels" and color to 'blue'.</li>
<li>Create ax2, as a twin of the first Axes.</li>
<li>In ax2, plot temperature against time, setting the color ylabel to "Relative temp (Celsius)" and color to 'red'.</li>
<li>Annotate the data using the ax2.annotate method. Place the text ">1 degree" in x=pd.Timestamp('2008-10-06'), y=-0.2 pointing with a gray thin arrow to x=pd.Timestamp('2015-10-06'), y = 1.</li>
</ul>

```python
fig, ax = plt.subplots()

# Plot the CO2 levels time-series in blue
plot_timeseries(ax, climate_change.index, climate_change['co2'], 'blue', "Time (years)", "CO2 levels")

# Create an Axes object that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 'red', "Time (years)", "Relative temp (Celsius)")

# Annotate point with relative temperature >1 degree
ax2.annotate(">1 degree", xy=(pd.Timestamp('2015-10-06'), 1), xytext=(pd.Timestamp('2008-10-06'), -0.2), arrowprops={"arrowstyle":"->", "color":"gray"})

plt.show()
```

## 3. Quantitative comparisons and statistical visualizations

### Bar chart
Instructions:
<ul>
<li>Call the ax.bar method to plot the "Gold" column as a function of the country.</li>
<li>Use the ax.set_xticklabels to set the x-axis tick labels to be the country names.</li>
<li>In the call to ax.set_xticklabels rotate the x-axis tick labels by 90 degrees by using the rotation key-word argument.</li>
<li>Set the y-axis label to "Number of medals".</li>
</ul>

```python
fig, ax = plt.subplots()

# Plot a bar-chart of gold medals as a function of country
ax.bar(medals.index, medals["Gold"])

# Set the x-axis tick labels to the country names
ax.set_xticklabels(medals.index, rotation=90)

# Set the y-axis label
ax.set_ylabel("Number of medals")

plt.show()
```

### Stacked bar chart
Instructions:
<ul>
<li>Call the ax.bar method to add the "Gold" medals. Call it with the label set to "Gold".</li>
<li>Call the ax.bar method to stack "Silver" bars on top of that, using the bottom key-word argument so the bottom of the bars will be on top of the gold medal bars, and label to add the label "Silver".</li>
<li>Use ax.bar to add "Bronze" bars on top of that, using the bottom key-word and label it as "Bronze".</li>
</ul>

```python
fig, ax = plt.subplots()
# Add bars for "Gold" with the label "Gold"
ax.bar(medals.index, medals["Gold"], label="Gold")

# Stack bars for "Silver" on top with label "Silver"
ax.bar(medals.index, medals["Silver"], bottom=medals["Gold"], label="Silver")

# Stack bars for "Bronze" on top of that with label "Bronze"
ax.bar(medals.index, medals["Bronze"], bottom=medals["Gold"] + medals["Silver"], label="Bronze")

# Display the legend
ax.legend()

plt.show()
```

### "Step" histogram
Instructions:
<ul>
<li>Use the hist method to display a histogram of the "Weight" column from the mens_rowing DataFrame, label this as "Rowing".</li>
<li>Use hist to display a histogram of the "Weight" column from the mens_gymnastics DataFrame, and label this as "Gymnastics".</li>
<li>For both histograms, use the histtype argument to visualize the data using the 'step' type and set the number of bins to use to 5.</li>
<li>Add a legend to the figure, before it is displayed.</li>
</ul>

```python
fig, ax = plt.subplots()

# Plot a histogram of "Weight" for mens_rowing
ax.hist(mens_rowing["Weight"], histtype='step', bins=5, label='Rowing')

# Compare to histogram of "Weight" for mens_gymnastics
ax.hist(mens_gymnastics["Weight"], histtype='step', bins=5, label='Gymnastics')

ax.set_xlabel("Weight (kg)")
ax.set_ylabel("# of observations")

# Add the legend and show the Figure
ax.legend()
plt.show()
```
